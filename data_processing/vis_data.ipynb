```python
hit_path = "C:/Users/arwin/Documents/dev/APS360-PROJECT/data/billboard_spec/Kanye West_Come To Life.png"
miss_path = "C:/Users/arwin/Documents/dev/APS360-PROJECT/data/random_spec/Rudy Kalma_Center.png"
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
hit_img = Image.open(hit_path)
miss_img = Image.open(miss_path)
```

```python
img_array = np.array(hit_img)

# Visualize the original image
plt.figure(figsize=(10, 4))
plt.subplot(1, 4, 1)  # 1 row, 4 columns, 1st subplot
plt.imshow(img_array)
plt.title('Spectrogram of "Come to Life"\nby Kanye West (Billboard Hit)')
plt.axis('off')  # Hide the axis

# Split the channels
r, g, b = img_array[:, :, 0], img_array[:, :, 1], img_array[:, :, 2]

# Visualize each channel independently
for i, channel in enumerate([r, g, b], start=2):  # Start from the 2nd subplot
    plt.subplot(1, 4, i)
    plt.imshow(channel, cmap='gray')  # Display in grayscale to show intensity
    plt.title(f'Channel {i-1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
